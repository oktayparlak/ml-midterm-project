```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.metrics import confusion_matrix
```

```python
label_df = pd.read_csv('labels.csv')
data_df = pd.read_csv('data.csv')
data_df.head()
```

```python
merged_df = pd.merge(label_df, data_df, left_on='Sample', right_on='Unnamed: 0').drop(columns='Unnamed: 0')
```

```python
X = merged_df.drop(['Sample', 'disease_type'], axis=1)
y = merged_df['disease_type']
# X_normalized = X.div(X.sum(axis=1), axis=0)
# X_train, X_test, y_train, y_test = train_test_split(X_normalized, y, test_size=0.20, random_state=42)
# model = LogisticRegression()
# model.fit(X_train, y_train)
# y_pred = model.predict(X_test)
# accuracy = accuracy_score(y_test, y_pred)
```

```python
test_sizes = [0.53, 0.77, 0.41]
for test_size in test_sizes:
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=test_size, random_state=42)


    gb_classifier = GradientBoostingClassifier(n_estimators=100, random_state=42)
    gb_classifier.fit(X_train, y_train)


    gb_accuracy = gb_classifier.score(X_test, y_test)
    gb_conf_matrix = confusion_matrix(y_test, gb_classifier.predict(X_test))


    gb_sensitivity = gb_conf_matrix[0, 0] / (gb_conf_matrix[0, 0] + gb_conf_matrix[0, 1])
    gb_specificity = gb_conf_matrix[1, 1] / (gb_conf_matrix[1, 0] + gb_conf_matrix[1, 1])
```
